```python
#Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?
```

```python
#Importando bibliotecas
import pandas as pd
```

```python
#Abrindo arquivo CSV
data_set = pd.read_csv("dataset.csv")
data_set.head()
```

```python
#Separando a categoria Office Supplies
data_set_office = data_set[data_set["Categoria"]== "Office Supplies"]

#Agrupando somátorio de Valor_Venda por cidade e categoria
data_set_city_office_venda = data_set_office.groupby("Cidade")["Valor_Venda"].sum().reset_index()

#filtrando por maior valor
data_set_city_office_venda_max = data_set_city_office_venda.max()
data_set_city_office_venda_max
```

```python
data_set_city_office_venda.sort_values(by = "Valor_Venda", ascending = False)
```
