```python
#Importando bibliotecas
import pandas as pd
import datetime as dt
```

```python
#Abrindo arquivo CSV
data_set = pd.read_csv("dataset.csv")
data_set.head()
```

```python
#Convertendo Data_Pedido em datetime
data_set["Data_Pedido"] = pd.to_datetime(data_set["Data_Pedido"], dayfirst = True)
data_set.dtypes
```

```python
#Extraindo o ano de Data_Pedido
data_set["Ano"] = data_set["Data_Pedido"].dt.year
data_set.head()
```

```python
#Agrupando Ano e Segmento por Valor_Venda
data_set_segmento_year_valor = data_set.groupby(["Segmento", "Ano"])["Valor_Venda"].sum()
data_set_segmento_year_valor
```
