```python
#Qual o Total de Vendas por Estado? Demonstre o resultado através de um gráfico de barras.
```

```python
#Importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Abrindo arquivo CSV
data_set = pd.read_csv("dataset.csv")
data_set.head()
```

```python
#Agrupando somátorio de Vendas por Estado
data_set_estado_venda = data_set.groupby("Estado")["Valor_Venda"].sum().reset_index()
data_set_estado_venda.head()
```

```python
#Definindo eixo_x e eixo_y
eixo_x = data_set_estado_venda["Estado"]
eixo_y = data_set_estado_venda["Valor_Venda"]

#Definindo tamanho da plotagem
plt.figure(figsize = (17, 6))

#Contruindo gráfico
sns.barplot(data = data_set_estado_venda, x = eixo_x, y = eixo_y).set(title = "Total de Vendas por Estado")
plt.xticks(rotation = 75)
plt.show()
```
