```python
'''Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. 
Qual Seria a MÃ©dia do Valor de Venda Antes e Depois do Desconto?'''
```

```python
#Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
#Abrindo arquivo CSV
data_set = pd.read_csv("dataset.csv")
data_set.head()
```

```python
#Criando nova coluna com desconto 15%
data_set["Desconto"] = np.where(data_set["Valor_Venda"] > 1000, 0.15, 0.10)
data_set.head()
```

```python
#Criando coluna com o desconto aplicado.
data_set["Valor_Com_Desconto"] = data_set["Valor_Venda"] - data_set["Valor_Venda"] * data_set["Desconto"]
data_set.head()
```

```python
#Filtrando colunas com desconto igual a 15%
Valor_Venda = data_set.loc[data_set["Desconto"] == 0.15, "Valor_Venda"]
Valor_Com_Desconto = data_set.loc[data_set["Desconto"] == 0.15, "Valor_Com_Desconto"]

#Tirando a media dos valores
Media_Valor_Venda = Valor_Venda.mean()
Media_Valor_Com_Desconto = Valor_Com_Desconto.mean()
```

```python
#exibindo resultados
print(f'media das vendas sem desconto:{Media_Valor_Venda}, media das vendas com desconto:{Media_Valor_Com_Desconto}')
```
